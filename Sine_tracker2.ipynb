```
# -*- coding: utf-8 -*-
"""
Created on Fri Feb 26 18:04:32 2021

@author: Max Parker
"""
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plot

# parameters of the tracking program
Fs = 60                                     # sample frequency
Ts = 1/Fs                                   # sample time
duration = 60.0                             # trial duration
time = np.arange(0, duration/Ts, 1);        # time
f = .15                                      # sine frequency

# parameters for the reorganisation model
reorg_gain = .000003                        # gain of reorganisation system (learning rate parameter)
ROC = random.rand(3)-.5                     # initial rate of change parameters for trial and error learning
reorg_err_threshold = 200                   # reorganise only when error exceeds threshold
err_memory = 50                            # duration of average error

# Amplitude of the target
target = 500*np.sin(2*np.pi*f*time*Ts)      # target position
target_vel = np.diff(target,1)              # target velocity
# plot.plot(time*Ts,target)

# Model parameters
theta = [500.00,.1,0.00]                   # sine input function init conditions
tau = 10                                   # input delay
K = 30                                     # output gain
S = .1                                     # leak rate
R = 0                                      # Reference value

cursor = np.zeros(int(duration/Ts))        # cursor init position
cursor_vel = np.zeros(int(duration/Ts))    # cursor init velocity

memory_trace=np.zeros(2)                   # initialise memory
e=np.zeros(int(duration/Ts))               # initialise error
abs_e=np.zeros(int(duration/Ts))           # initialise absolute error
s=[int(duration/Ts),3]                                        
parameters=np.zeros(s)                     # initialise parameter list
running_av_err=np.zeros(int(duration/Ts))  # initialise running_average_error for reorg



# model
for i in range(tau,int(duration/Ts)):
    
    
    ########### Track ############
    leng=np.arange(1, i, 1)
    if i == tau:
        memory_trace=target[i-tau]
    else:
        memory_trace=np.append(memory_trace,target[i-tau] + theta[2])          # laying memory of previously seen trace as it comes in
    
    pattern = theta[0]*np.sin(2*np.pi*theta[1]*leng*Ts + theta[2])     # fit sine to memory trace  
    
    x=np.arange(1,duration/Ts+1)
    pattern_est = theta[0]*np.sin(2*np.pi*theta[1]*x*Ts);           # extrapolate sine across length of trial  
    pred = pattern_est[i]                                           # predicted position of target location on basis of model
    

    error = R-(cursor[i-tau] - pred)                                # error signal
    cursor[i] = cursor[i-1] + (S*(K*(error) - cursor[i-1]))*Ts;     # output
    cursor_vel[i] = cursor[i]-cursor[i-1];                          # output velocity
    
    e[i] = cursor[i] - target[i]                                    # tracking error
    abs_e[i]= abs(e[i])                                             # absolute tracking error
    
    
    ########### Reorganise #########
    parameters[i,] = theta                                          # parameters on current trial
    if i > err_memory:
        running_av_err[i] = np.mean(abs_e[i-err_memory:i-tau])
        
        if running_av_err[i] > reorg_err_threshold:                 # reorganise if abs error is above reorg threshold
        
            e_dt = running_av_err[i] - running_av_err[i-1];         # rate of change in tracking error
    
            if (e_dt > 0):                                          # if error increased
                ROC = random.rand(3)-.5                             # set new rate of change
    
            theta = theta + reorg_gain * (running_av_err[i] * ROC); # reorganise
    
            if theta[0] < 0:
                theta[0] = parameters[i,0];                         # stop parameters going below zero

            if theta[1]< 0:
                theta[1] = parameters[i,1];                         # stop parameters going below zero
        else:        
            theta =  parameters[i,]  
    else:
        theta =  parameters[i,]  

    parameters[i,] = theta

plot.figure(1)
plot.plot(time*Ts,target,time*Ts,cursor)
plot.xlabel('Time (s)')
plot.ylabel('Position')
plot.legend(['Target','Cursor'])

plot.figure(2)
plot.plot(time*Ts,parameters[:,1])
plot.xlabel('Time (s)')
plot.ylabel('Reproduced Frequency Paremeter')

plot.figure(3)
plot.plot(time*Ts,running_av_err)
plot.xlabel('Time (s)')
plot.ylabel('Running Average Error')
```
